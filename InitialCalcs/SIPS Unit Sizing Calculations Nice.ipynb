# The Use of Solvent Impregnated Polymers in Unit Operations. 7/4/2017

## Introduction

Solvent Impregnated Polymers (SIPs - our working acronym :) ) are a newly developed material for gas absorption processes such as carbon capture. Micron-sized droplets of liquid solvent are dispersed inside a solid polymer material which is highly permeable to the solute gas. Potential benefits of this material include:

 - *SIPs are simple to manufacture.* A single solvent/polymer emulsion is created, and then, once this emulsion has been moulded into the desired shape (coated onto a surface, dispersed as small droplets in a double emulsion, etc.) the polymer is cured (currently via UV light.) SIPs were inspired by microencapsulation of solvents (MECS), in which liquid solvents are encapsulated in small polymer shells (with diameter of 100-500 microns). However, MECS are made at g/h scale using microfluidic devices, while SIPs can be created using bucket chemistry.
 - *Enhanced mass transfer*. In many gas absorption operations, including many carbon capture operations, absorption is controlled by diffusion on the liquid side. Thus, any increase in gas-liquid surface area will proportionally increase the gas absorption rate. This was the inspiration for MECS, however mass transfer modelling shows that even microencapsulated solvents with diameter of a few 100 microns still have Thiele Modulii much larger than 1, and so would benefit from further increases in the solvent's specific surface area. Mass transfer into SIPs is substantially faster than into the (already impressive) MECS, and this is achieved by taking advantage of the unintuitive fact that many polymers have exceptionally high gas permeabilities. To take a specific example, the permeability of CO2 in PDMS is over an order of magnitude higher than many liquid solvents such as concentrated carbonate solutions or ionic liquids. CO2 can travel through the PDMS network, and react in a (reaction-controlled) manner with the micron-sized droplets of liquid solvent embedded within it. PDMS provides the high permeability and maintains the high solvent surface area, while the solvent provides the selectivity and capacity. MECS of size ~100 microns have 1-2 orders of magnitude greater surface area than solvents in a packed column; SIPs have 1-2 orders of magnitude greater solvent surface area again (though of course, this must be qualified by noting that gas must still diffuse through the polymer.) Simple models like those developed below show that a 5-fold improvement in mass transfer over MECS can be expected.
 - *Imobilisation of liquid solvent*. Many solvents for CCS are too viscous, volatile or corrrosive to be easily used in traditional processing equipment. For instance, some protic ionic liquids have been shown to have comparable CO2 capacities to amines like MEA, yet can be regenerated at 55C. But these liquids are viscous, leading to poor mass transfer and handling difficulties. The SIPs motif could solve both of these problems.
 - *Flexibility in designing gas-liquid contactors*. The fact that SIPs are manufactured by crosslinking a single emulsion opens up some interesting possiblities for gas-liquid contacting. They could be coated on the sides of a monolith through which flue gases can flow with minimal pressure drop. The simplest design would be a checkerboard pattern: flue gases pass through one set of channels (the 'white' squares on the checkerboard), and then, once the SIP is saturated, hot gases could pass through the adjacent ('black') squares: heat passing through the walls and releasing the stored gas. In combination with the PIL's described above, this could lead to processing units with minimal pressure drops, and which could utilise waste heat streams to regenerate the solvent at little cost. SIPs could be shaped into droplets of varying size, for use in fluidised or packed beds.
 
It's clear the SIPs motif has a lot of potential. There are some possible pitfalls ahead of us:
 - Can we stabilise the solvet/polymer emulsion long enough for the SIPs to be moulded into the desired shape before polymerisation occurs. Initial experiments are promising, but more systematic work is required.
 - How high can we make the volume fraction of solvent? If it is too high, phase inversion will occur in the single emulsion.
 - As capital costs are likely to be high, is this material stable long-term? Can polymers and solvents be found that are stable to SOx/NOx poisoning, etc.
 - Can we design unit operations that utilise this material in a competitive way?

Below we analyse the last question. My main focus in what follows will be on the monolith design described above. Packed bed arrangements (with particles supported in larger, permable filler materials) or fluidised bed arrangements will be discussed later.

## Model for Absorption into SIPs Material

### Discussion

The rigorous modelling of solvent impregnated polymers is a daunting task. Diffusion and reaction occurs at multiple length scales. If embedded solvent droplets are large enough, even the absorption of gas into spherical particles or flat surfaces of SIPs could become a multi-dimensional problem. For our initial model, we'll make the following simplifications:
 - The diameter of solvent droplets is much smaller than then length-scale over which concentrations in the PDMS change. Thus the material can be viewed as continuous from the point of view of modelling changes in the concentration profile over time.
 - Solvent droplets are so small that their dynamics are entirely reaction controlled (the Thiele Modulus is much less than 1). Combining this with the assumption that the establishment of equilibrium at the polymer-solvent interface  is a very fast process, we can say that the concentration of gas in the polymer and in the solvent are related by their relative solubilities: $c_{solv}/\mathcal{S}_{solv} = c_{poly}/\mathcal{S}_{poly}$. Here $c$ is the solubility of gas in mol/m3, and $\mathcal{S}$ is the solubility, in mol/Pa.m3.
 - The permeability of the gas through the polymer is much higher than in the liquid, so we consider that across any slice normal to a macroscopic concentration gradient, gas only flows through the PDMS phase.
 - The solvent reacts with the gas according to a pseudo first-order reaction, with reaction-rate constant $k$. This is an excellent assumption for many solvents, however this pseudo-first order reaction rate constant will change over time (as the reactive species in the solvent is exhausted) and this will happen faster near the gas-SIP boundary, where concentrations are higher during absorption. For now we ignore this, and consider constant $k$. This is a good model for the initial rate of absorption. The time and spatial dependence of $k$ as the SIP becomes saturated will be considered at a later date.
 
### Mass Balance

With all of these assumptions, we can formulate a diffusion-reaction equation in a consistent way. To model flow into a flat wall of SIP of thickness $\delta$, we consider a thin flat shell, normal to the gas-SIP surface, with unit area and thickness $\Delta x$. For this geometry:

- The build up of gas inside the polymer phase is given by 
$$
\Delta x \varepsilon \frac{\partial c_{poly}}{\partial t} 
$$
here $\varepsilon$ is polymer fraction in the SIP. The build up inside the solvent phase is
$$
\Delta x (1-\varepsilon) \frac{\partial c_{solv}}{\partial t} = \Delta x (1-\varepsilon) \frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}} \frac{\partial c_{poly}}{\partial t}.
$$
The total rate of build up of gas is then:
$$
\left( \varepsilon +  (1-\varepsilon) \frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}} \right) \frac{\partial c_{poly}}{\partial t} \Delta x
$$

- The net diffusive flow of gas into the shell via the two walls (each with unit area, and given that we're ignoring diffusive flux in the solvent phase) is:

$$
\mathcal{D}_{poly}\varepsilon \left(-\frac{\partial c_{poly}}{\partial x}\Big|_{x} - -  \frac{\partial c_{poly}}{\partial x}\Big|_{x+\Delta x} \right)
$$

- The net rate of reaction of gas inside the liquid solvent is given by
$$
kc_{solv}(1-\varepsilon)\Delta x = k\frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}}c_{poly}(1-\varepsilon)\Delta x 
$$

Overall, then, the mass balance gives:

$$
\left( \varepsilon +  (1-\varepsilon) \frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}} \right) \frac{\partial c_{poly}}{\partial t} \Delta x = \mathcal{D}_{poly}\varepsilon \left(-\frac{\partial c_{poly}}{\partial x}\Big|_{x} - -  \frac{\partial c_{poly}}{\partial x}\Big|_{x+\Delta x} \right) -  k\frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}}c_{poly}(1-\varepsilon)\Delta x 
$$

Taking $\Delta x \rightarrow 0$, we have
$$
\left( \varepsilon +  (1-\varepsilon) \frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}} \right) \frac{\partial c_{poly}}{\partial t}  = \mathcal{D}_{poly}\varepsilon \frac{\partial^2 c_{poly}}{\partial x^2} -  k\frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}}c_{poly}(1-\varepsilon)
$$

### Simplifying the PDE

Setting $c\equiv c_{poly}$, we can rearrange this to a familiar-looking pde:
$$
\frac{\partial c}{\partial t}  = \mathcal{D}' \frac{\partial^2 c}{\partial x^2} -  k'c
$$

with BC's 
$$c|_{x=\delta} = p_0S_{poly}$$ 
and 
$$\partial_x c|_{x=0} = 0.$$

We have:

$$
\mathcal{D}' = \mathcal{D}_{poly}\varepsilon \Bigg/\left( \varepsilon +  (1-\varepsilon) \frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}} \right)
$$

and 

$$
k' = k\frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}}(1-\varepsilon)\Bigg/ \left( \varepsilon +  (1-\varepsilon) \frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}} \right)
$$

We note in passing that for spherical SIPs, the modified PDE is applicable:

$$
\frac{\partial c}{\partial t}  = \mathcal{D}' \frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2\frac{\partial c}{\partial r}\right) -  k'c
$$
with BC's 
$$c|_{r=R} = p_0S_{poly}$$ 
and 
$$\partial_r c|_{r=0} = 0$$. 

### Solving the PDE

Returning to the linear case, it is well known that the solution to this pde quickly assymptotes to a steady-state concentration profile given by:

$$
c = \mathcal{S}_{\text{poly}} p_0 \frac{\cosh(\phi x/\delta)}{\cosh(\phi)}
$$

where

$$
\phi^2 = \delta^2\frac{k'}{\mathcal{D}'} = \delta^2 k\frac{\mathcal{S}_{solv}}{\mathcal{S}_{poly}}(1-\varepsilon) \Bigg/\mathcal{D}_{poly}\varepsilon = \delta^2 \frac{(1-\varepsilon)k\mathcal{S}_{solv}}{\varepsilon\mathcal{P}_{poly}}$$

Now, the flux into the surface is given by:

$$
\text{Flux} = \mathcal{D}_{\text{poly}}\varepsilon \frac{\partial c}{\partial x}\Big|_{x=\delta}
$$

And so
$$
\text{Flux} = \frac{\mathcal{P}_{\text{poly}} p_0 \varepsilon \phi}{\delta} \tanh(\phi)
$$

This is the expression we will use for our models of flow into a wall of SIPs. Before we do that, however, we'll take a slight detour to see if this expression makes physical sense in various assymptotic limits.

### Assymptotic Analysis

We can see how this flux expression simplifies when $\phi << 1$ (reaction controlled) and $\phi >> 1$ (diffusion controlled). In the former, $\tanh(\phi) \approx \phi$, so 

$$
\text{Flux} \approx \frac{\mathcal{P}_{\text{poly}} p_0 \varepsilon }{\delta}\phi^2 = \frac{\mathcal{P}_{\text{poly}} p_0 \varepsilon}{\delta} \left(\delta^2 \frac{(1-\varepsilon)k\mathcal{S}_{solv}}{\varepsilon\mathcal{P}_{poly}}\right) = k\mathcal{S}_{solv} p_0(1-\varepsilon)\delta
$$

which is exactly what would be expected if reaction in the solvent phase was the rate-limiting factor. For CCS applications, where solvents are typically diffusion controlled, if we can get our material operating in this regime, this would be a huge success. More realistically, we'll probably still be operating with $\phi >> 1$, at which point we are still diffusion-controlled, and $\tanh(\phi) \approx 1$. Then

$$
\text{Flux} \approx \frac{\mathcal{P}_{\text{poly}} p_0 \varepsilon }{\delta} \phi = \frac{\mathcal{P}_{\text{poly}} p_0 \varepsilon}{\delta} \left(\delta \sqrt{\frac{(1-\varepsilon)k\mathcal{S}_{solv}}{\varepsilon\mathcal{P}_{poly}}}\right) = p_0 \sqrt{\varepsilon(1-\varepsilon)\mathcal{P}_{poly}\mathcal{S}_{solv}k} = p_0 \mathcal{S}_{solv} \sqrt{\frac{\varepsilon(1-\varepsilon)\mathcal{P}_{poly}k}{\mathcal{S}_{solv}}}
= p_0 \mathcal{S}_{solv} \sqrt{\mathcal{D}_{solv}k} \left(\sqrt{\varepsilon(1-\varepsilon)\frac{\mathcal{P}_{poly}}{\mathcal{P}_{solv}}}\right)
$$

we write this in this form, because the diffusion controlled flux into a stationary, pure liquid (as would be found inside an MECS) is:

$$
p_0 \mathcal{S}_{solv} \sqrt{\mathcal{D}_{solv}k}
$$

and so the enhancement in the *flux* that SIPs are providing, compared to the static fluid, is given by:

$$
\sqrt{\varepsilon(1-\varepsilon)\frac{\mathcal{P}_{poly}}{\mathcal{P}_{solv}}}
$$

We take a hit because of $\varepsilon$ (for two reasons: diffusive flux only occurs in the polymer, and reaction only occurs in the solvent) but if the polymer has much higher permeability, we can get greater flux. Remember this is the enhancement in the *flux* into SIPs, compared with into stationary fluids. But total flow is flux multiplied by surface area. If you combine this higher flux with the fact that, compared to free fluids, SIPs can be moulded into shapes with much higher specific surface areas (such as small particles or thin walls on monoliths) the potential improvement in overall mass transfer increases even more.

## Model For Flow of Gas through SIP-coated Monolith with Square Tubes. 

### Assumptions

We're going to make some simplifying assumptions in order to create this model:
 - First, we'll use the flux expression derived above as the basis for our calculations, so every assumption built into that model will be used here too. In particular, we do not consider changes in $k$ over time or space. In reality, $k$ will vary in both the direction of the gas flow, and into the material perpendicular to the gas flow. The actual mass transfer zone that we establish will be longer than calculated here - this is an upper limit. 
 - The fact that we are using the steady-state model above also means we are assuming that the gas-phase concentration is changing slowly, so the SIP has time to reach its steady-state solution locally. This is a reasonable assumption.
 - We ignore axial migration of gas inside the SIP.
 - We consider the walls of the monolith to be very thin, so their volume is negligible. However, we will account for the SIP volume: the more SIP, the more the pores will be filled, and the less surface area will be available for gas transfer. 
 - We will assume that the 'corners' of each square cell, which will be filled with SIP, do not affect the flux into the SIP, though they do store gas. This will mean we will slightly underestimate the gas flux.

### Model

Suppose that each cell of the monolith has width $w$, so the total number of cells per unit area is $1/w^2$. Suppose that on each of the four walls of every cell there is a layer of SIP of thickness $\delta$. Then void space in a unit volume of monolith (or the monolith void fraction, not to be confused with the polymer fraction in the SIP) will be given by $$\xi=1 - \frac{4(w-\delta)\delta}{w^2}.$$

Clearly we'd expect $0<\xi<1$, and this requires $0<\delta < w/2$, which is physically reasonable.

The surface area of SIP being exposed to the gas in a unit  volume of monolith is

$$
a = \frac{4(w-2\delta)}{w^2}
$$

Meanwhile, from above, the flux of gas into the SIP is given by

$$
\mathcal{J} = \frac{\mathcal{P}_{\text{poly}} p \varepsilon \phi}{\delta} \tanh(\phi)
$$

where

$$
\phi = \delta \sqrt{\frac{(1-\varepsilon)k\mathcal{S}_{solv}}{\varepsilon\mathcal{P}_{poly}}}
$$

and $p$ is the partial pressure of the gas being absorbed at that location in space. We are now in a position to do a shell balance over a small section of the monolith.

### Monolith Shell Balance

Consider a monolith of unit cross-sectional area, and consider a thin slice at position $z$ and of width $\Delta z$. Gas flows through the open fraction $\xi$ of the monolith, and absorbs into the SIP coating the wall. The flux of gas into the wall is given by the expression above. We will assume that the total superficial gas velocity, $u$, is a constant; this is reasonable for dilute gas flows, and will only give a small error for carbon capture, in which $x_{CO2} \approx 10\%$. Then for our thin slice:
 - We'll assume we're operating at steady-state, so there is no dynamic term in the pde.
 - The net flow of gas in and out via the tubes is given by:
 $$ \frac{u}{RT}p|_{z} - \frac{u}{RT}p|_{z+\Delta z}$$
 - The net flow of gas into the SIP is given by
 $$ \mathcal{J} \times A = \left(\frac{\mathcal{P}_{\text{poly}} \varepsilon \phi}{\delta} \tanh(\phi)\right) p \times \frac{4(w-2\delta)}{w^2} \Delta z$$
 where

$$
\phi = \delta \sqrt{\frac{(1-\varepsilon)k\mathcal{S}_{solv}}{\varepsilon\mathcal{P}_{poly}}}
$$

The overall mass balance is then given by

$$
0 = \frac{u}{RT}p|_{z} - \frac{u}{RT}p|_{z+\Delta z} - \left(\frac{\mathcal{P}_{\text{poly}} \varepsilon \phi}{\delta} \tanh(\phi) \times \frac{4(w-2\delta)}{w^2} \right) p \Delta z
$$

And in the limit as $\Delta z \rightarrow 0$,
$$
\frac{dp}{dz} = - \frac{RT}{u}\left(\frac{\mathcal{P}_{\text{poly}} \varepsilon \phi}{\delta} \tanh(\phi) \times \frac{4(w-2\delta)}{w^2} \right) p = -\alpha p
$$

And so

$$
p(z) = p^*\exp \left(-\frac{\alpha}{u} z\right)
$$

where $p^* = p|_{z=0}$, and 

$$
\alpha = \frac{4(w-2\delta)}{w^2}\frac{\mathcal{P}_{\text{poly}} \varepsilon \phi}{\delta} RT\tanh(\phi). 
$$

The length of monolith required to remove some fraction $\Gamma$ of the gas is:

$$
L_{\Gamma} = -\frac{\ln 1-\Gamma}{\alpha}u
$$

### Analysis of Operating Unit

Notice that 

$$
\frac{L_{\Gamma}}{u} = -\frac{\ln 1-\Gamma}{\alpha} \equiv \Omega = \text{Operational Constant} 
$$

meaning that if we changed the volume of the reactor, or the gas flow rate, this term wouldn't change significantly.

Furthermore, we have a specific amount of gas to process, so 

$$
u\times A = \text{Total Volumetric Gas Flow Rate} \equiv Q
$$

Now, it is clear that the total volume of the reactor required by the kinetics of the absorption is given by:

$$
V = L_{\Gamma}A = \Omega Q
$$

We shall now code this up, insert some real values, and see what sort of reactor volumes are required in order to absorb 90\% of the CO2 from a flue gas stream

## Function for Reactor Volume

```julia
#Calculate the total volume of a monolith removing a given fraction of gas.
function MonolithVolume(P_poly,S_sol,ɛ,k,δ,CPSI,R,T,Γ,Q)
    w = 1./sqrt(CPSI*1550)   #CPSI is cells per square inch. w is the cell side length.
    ϕ = δ.*sqrt(((1-ɛ).*k.*S_sol) ./ (ɛ.*P_poly))
    α = 4*(w-2*δ)./w.^2.*P_poly.*ɛ.*ϕ./δ.*R.*T.*tanh(ϕ)
    Ω = -log(1-Γ)./α
    V = Ω.*Q
    return V
end
```
